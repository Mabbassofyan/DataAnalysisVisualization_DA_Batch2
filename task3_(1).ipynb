```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('imdb_top_1000.csv_cleaned.csv')
```

**Understanding the distribution of ratings**

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['IMDB_Rating'], bins=20, kde=True)
plt.title('Distribution of IMDB Ratings')
plt.xlabel('IMDB Rating')
plt.ylabel('Number of Movies')
plt.show()
```

 **Identifying the most frequent genres**

```python
genre_counts = data['Genre'].str.split(', ').explode().value_counts()
top_genres = genre_counts[:10]  # Select the top 10 genres

plt.figure(figsize=(12, 6))
sns.barplot(x=top_genres.index, y=top_genres.values)
plt.title('Top 10 Movie Genres')
plt.xlabel('Genre')
plt.ylabel('Number of Movies')
plt.xticks(rotation=45, ha='right')
plt.show()
```

**Exploring if higher-rated movies tend to earn more**

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='IMDB_Rating', y='Gross', data=data)
plt.title('Correlation between Gross and IMDB Rating')
plt.xlabel('IMDB Rating')
plt.ylabel('Gross (in billions)')
plt.show()
```
