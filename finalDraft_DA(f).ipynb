# **Building the Regression Model**

**Problem**: Predicting IMDB rating based on gross and runtime.

**Model**: Linear Regression

**Data Preparation**: Handled missing values, scaled features.

**Split**: Stratified split (80% train, 20% test).

**Correlation**: Analyzed using heatmap.

**Accuracy**: Evaluated using MSE and R-squared.

**Visualization**: Scatter plot of actual vs. predicted ratings, Scatter Plots of Features vs. Target, and Residual Plot.

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns


data = pd.read_csv('/content/drive/MyDrive/DataAnalysisVisualization_DA_Batch2-main/imdb_top_1000.csv_cleaned.csv')

# Extract numeric values from 'Runtime' column
data['Runtime'] = data['Runtime'].str.extract('(\d+)').astype(float)

# Select features and target variable
X = data[['Gross', 'Runtime']]
y = data['IMDB_Rating']

# Handle missing values (if any)
X.fillna(X.mean(), inplace=True)
```

**Stratified Split**

```python
from sklearn.model_selection import StratifiedShuffleSplit

# Create bins for IMDB_Rating for stratification
bins = pd.qcut(y, q=5, labels=False)  # Create 5 quantile-based bins

#We use StratifiedShuffleSplit to ensure the train and test sets have similar distributions of IMDB ratings.
splitter = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
train_index, test_index = next(splitter.split(X, bins))

X_train, X_test = X.iloc[train_index], X.iloc[test_index]
y_train, y_test = y.iloc[train_index], y.iloc[test_index]

# Scale the features using StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

**Correlation Analysis**

```python
correlation_matrix = data[['Gross', 'Runtime', 'IMDB_Rating']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

**Building the Model**

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

**Accuracy Evaluation**

Lower MSE and higher R-squared indicate better performance.

Great question! The quality of a Mean Squared Error (MSE) value really depends on the context. An MSE of 0.079 might be fantastic in some situations and underwhelming in others. Let me break it down for you:

Comparative Analysis: If you're comparing this model to others, an MSE of 0.079 might be lower, implying it's performing better.

Scale of Dependent Variable: If your dependent variable's values range from, say, 0 to 1, an MSE of 0.079 is relatively small. However, if they range from 0 to 100, the same MSE could be more significant.

Ultimately, an MSE value needs to be evaluated in relation to your specific dataset and domain of application.

```python
y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")
```

**Visualization**

```python
#Distribution of Target Variable (IMDB Rating)
plt.scatter(y_test, y_pred)
plt.xlabel('Actual IMDB Rating')
plt.ylabel('Predicted IMDB Rating')
plt.title('Actual vs. Predicted IMDB Rating')
plt.show()
print('\n')
print('\n')

#Scatter Plots of Features vs. Target
plt.figure(figsize=(10, 6))
sns.histplot(data['IMDB_Rating'], bins=20, kde=True)
plt.title('Distribution of IMDB Ratings')
plt.xlabel('IMDB Rating')
plt.ylabel('Frequency')
plt.show()
print('\n')
print('\n')

#Residual Plot
plt.figure(figsize=(10, 6))
sns.residplot(x=y_pred, y=y_test)
plt.title('Residual Plot')
plt.xlabel('Predicted IMDB Rating')
plt.ylabel('Residuals')
plt.show()
```

# **Building the Classification Model**

**Problem**: Classifying IMDB rating as "High" or "Low" based on gross and runtime.

**Model**: Logistic Regression

**Data Preparation**: Created a binary target variable, handled missing values, scaled features.

**Split**: Stratified split (80% train, 20% test).

**Correlation**: Skipped for simplicity in this classification example.

**Accuracy**: Evaluated using accuracy score and classification report.

**Visualization**: Confusion matrix.

**Data Preparation**

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('/content/drive/MyDrive/DataAnalysisVisualization_DA_Batch2-main/imdb_top_1000.csv_cleaned.csv')

# Create a binary target variable for classification
data['Rating_Category'] = pd.qcut(data['IMDB_Rating'], q=2, labels=['Low', 'High'])

# Select features and target variable
X = data[['Gross', 'Runtime']]
y = data['Rating_Category']

# Handle missing values (if any)
# Extract numeric values from 'Runtime' column before filling NaNs
X['Runtime'] = X['Runtime'].str.extract('(\d+)').astype(float)
X.fillna(X.mean(), inplace=True) # Now fill NaNs with the mean of numeric values
```

**Stratified Split**

```python
from sklearn.model_selection import StratifiedShuffleSplit

# Perform stratified split
splitter = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
train_index, test_index = next(splitter.split(X, y))  # Stratify based on 'Rating_Category'

X_train, X_test = X.iloc[train_index], X.iloc[test_index]
y_train, y_test = y.iloc[train_index], y.iloc[test_index]

# Scale the features using StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

Correlation analysis can be adapted for categorical target variables For example, you could use point-biserial correlation for continuous features and a binary target. However, for simplicity, we'll skip it in this classification example

**Building the Logistic Regression Model**

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

**Accuracy Evaluation**

```python
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

print(f"Accuracy: {accuracy}")
print(report)
```

**Visualization**

```python
# Confusion Matrix
from sklearn.metrics import confusion_matrix
import seaborn as sns

cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Rating_Category', y='Gross', data=data)
plt.title('Gross by Rating Category')
plt.xlabel('Rating Category')
plt.ylabel('Gross')
plt.show()
print('\n')
print('\n')

plt.figure(figsize=(10, 6))
sns.boxplot(x='Rating_Category', y='Runtime', data=data)
plt.title('Runtime by Rating Category')
plt.xlabel('Rating Category')
plt.ylabel('Runtime')
plt.show()
```
